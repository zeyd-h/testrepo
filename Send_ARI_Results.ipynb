```python
import openpyxl  
def update_sheet(xlsxloc,lst):
    wb = openpyxl.load_workbook(xlsxloc)
#     sheet = wb.active
    sheets = wb.sheetnames
    sheet = wb["Data"]

    for i in range(len(lst)):
        if lst[i]==1:
            print(str(i+1)+": "+str(lst[i]))
            sheet.cell(row=i+2, column=30).value =int(lst[i])
    try:
        wb.save(xlsxloc)
    except:
        print("+++++++ could not save the file.")

    wb.close() 
    
```
